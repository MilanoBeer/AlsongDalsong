```python
data = data.drop(labels='Unnamed: 0.1', axis=1)
data
```

```python
# 유튜브 링크 크롤링 시작 ! 2022.09.16 - 14:59:13

print(int(data.loc[0, 'now'])-1)

for i in range(int(data.loc[0, 'now'])-1, len(data)):

    if data.loc[i, 'videoId'] == 'None':        
        search = data.loc[i, 'artist_name'] + " " + data.loc[i, 'track_name']      
        
        url = "https://www.youtube.com/results?search_query=" + search + "&sp=EgIQAQ%253D%253D"
        response = requests.get(url).text
        
        idx = response.find('webCommandMetadata":{"url":"/watch?v')
        
        if idx == -1:
            data.loc[i, 'videoId'] = 'NO VIDEO'
            print('NO VIDEO', end = "  ===============  ")
            print(i, end = "  ===============  ")
            print(search)
        
        else:
            start = idx + 37
            end = start + 11
            result = response[start:end]
            data.loc[i, 'videoId'] = result
            
            
        if i % 1000 == 0:
            data.to_csv("musics.csv")
            
            
        if i % 200 == 0:
            print(i, end = "  ======  ")
            print(time.strftime('%Y.%m.%d - %H:%M:%S'), end = "  ======  ")
            print(result, end = "  ======  ")
            print(search)
    
        data.loc[0, 'now'] = i      
    
```
